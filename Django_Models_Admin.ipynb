# Django Models & Admin

Actividad: Crear 500 productos de prueba en Django de forma automatizada y trabajar con fixtures para exportar e importar la información.

## Objetivo
Aplicar la creación masiva de objetos con Django ORM, guardar esos datos en un fixture JSON y comprobar que pueden restaurarse correctamente en la base de datos.


## Paso 1. Importaciones y usuario

En esta parte se importa el modelo `ProductModel` y el modelo de usuario para asociar cada producto a un usuario existente.


````python
from ecommerce.models import ProductModel
from django.contrib.auth import get_user_model

User = get_user_model()
user = User.objects.first()

user
````

## Paso 2. Eliminar productos previos

Antes de crear los 500 registros, se eliminan los productos anteriores para trabajar exactamente con la cantidad solicitada.


````python
ProductModel.objects.all().delete()
ProductModel.objects.count()
````

## Paso 3. Creación masiva de 500 productos

Se utiliza `bulk_create` para insertar múltiples registros de forma eficiente.


````python
products = []

for i in range(1, 501):
    products.append(
        ProductModel(
            user=user,
            title=f"Producto {i}",
            price=100 + i,
            description=f"Descripción {i}"
        )
    )

ProductModel.objects.bulk_create(products)
````

## Paso 4. Verificación de registros creados

El resultado esperado es mostrar `500` productos en la base de datos.


````python
ProductModel.objects.count()
````

## Paso 5. Exportación a fixture JSON

Comando utilizado en terminal para generar el fixture:

```bash
docker compose exec web sh -c "python manage.py dumpdata ecommerce --indent 4 --format json > ecommerce/fixtures/products/500Products.json"
```


## Paso 6. Verificación del fixture

Para revisar que el archivo JSON sí tenga contenido, se puede usar el siguiente comando:

```bash
docker compose exec web sh -c "head -n 5 ecommerce/fixtures/products/500Products.json"
```


## Paso 7. Eliminación de productos para probar la restauración

Una vez creado el fixture, se eliminan los productos de la base de datos.


````python
ProductModel.objects.all().delete()
ProductModel.objects.count()
````

## Paso 8. Carga del fixture

Comando utilizado para restaurar los productos desde el archivo JSON:

```bash
docker compose exec web python manage.py loaddata ecommerce/fixtures/products/500Products.json
```


## Paso 9. Comprobación final

Después de cargar el fixture, la base de datos debe volver a contener 500 productos.


````python
ProductModel.objects.count()
````

## Resultado esperado

Debe mostrar correctamente **500 productos** en la base de datos después de la restauración del fixture.


## Conclusión

En esta actividad se creó una colección de 500 productos utilizando creación masiva con `bulk_create`. Después, los datos se exportaron a un fixture JSON y se restauraron exitosamente con `loaddata`. Esto permite reutilizar información de prueba de manera rápida y organizada dentro de un proyecto Django.
